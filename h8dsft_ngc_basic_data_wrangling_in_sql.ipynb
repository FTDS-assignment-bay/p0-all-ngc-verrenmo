```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "mindful-carport-440904-q3"
client = bigquery.Client(project=project_id)
```

```python
# How many tables are in the austin_bikeshare dataset?
# 1. Mengambil informasi tabel dari database (namadatabase.INFORMATION_SCHEMA.TABLES)
# 2. Menghitung jumlah baris dari info tabel. Karena setiap tabel punya nama berbeda jadi jumlah baris = jumlah tabel
query = """
SELECT
  COUNT(table_name) as number_of_table
FROM
  `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES`

"""

query_job = client.query(query).to_dataframe()
query_job

# Ada 2 tabel dalam database austin_bikeshare
```

```python
# How many columns in the bikeshare_stations table have STRING data?
# 1. Mengambil informasi kolom dari database (namadatabase.INFORMATION_SCHEMA.COLUMNS)
# 2. Informasi yang diambil dengan kondisi nama tabel = bikeshare_stations dan memiliki tipe data = 'STRING'
# 3. Menghitung jumlah baris dari data yang sudah diambil dengan kondisi no.2
query = """
SELECT
  COUNT(table_name) as number_of_string_datatype
FROM
  `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE
  table_name = 'bikeshare_stations'
  AND
  data_type = 'STRING'

"""

query_job = client.query(query).to_dataframe()
query_job
```

```python
#  Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes.
# 1. Mengambil data di tabel bikeshare_trips
# 2. Filter menggunakan where dengan kondisi bike_id is not null dan duration_minutes>1000
query = """
SELECT
  *
FROM
  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE
  bike_id IS NOT NULL AND
  duration_minutes > 1000

"""

query_job = client.query(query).to_dataframe()
query_job
```

```python
# Filter trips where the subscriber_type is "Walk Up" and there's 'Street' or 'St.' in end_station_name .
# 1. Mengambil data dari tabel bikeshare_trips
# 2. Filter data dengan kondisi subscriber_type = 'Walk Up' dan end_station_name memiliki kata 'Street' or 'St.'
query = """
SELECT
  *
FROM
  `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE
  subscriber_type = 'Walk Up'
  AND
  (end_station_name LIKE '%Street%' or end_station_name LIKE '%St.%')



"""

query_job = client.query(query).to_dataframe()
query_job
```
