# Task 1 - Scrape Coin from Coin Market Cap

Scrape coins price from https://coinmarketcap.com/ with following requirements:

●Scrape up to 1000 coins

●Including Price, Market Cap, Volume, and Coin supply

●Save into Pandas DataFrame and then save into excel or csv file4


```python
driver = webdriver.Chrome()

# Menyiapkan list kosong untuk menyimpan data
listCoin = []
listPrice = []
listMarketCap = []
listVolume = []
listCoinSupply =[]


# Looping untuk mengambil data dari halaman pertama sampai ke 10 (1 halaman berisi 100 koin)
for i in range(1,11):
    # Membuka url halaman berdasarkan nomor halaman yang sedang di loop
    url = f"https://coinmarketcap.com/?page={i}"
    driver.get(url)

    # Melakukan scroll pada halaman untuk memuat seluruh konten
    for i in range(8):
        driver.execute_script("window.scrollBy(0,1080)")
        time.sleep(1)

    driver.execute_script("window.scrollBy(50,0)")
    time.sleep(1)

    # Mengambil data html dari halaman setelah scroll
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Menemukan tabel dan baris yang berisi data masing-masing koin
    tabel = soup.find('table', class_='sc-7b3ac367-3 etbcea cmc-table')
    body = tabel.find('tbody')
    rows = body.find_all('tr')

    # Looping untuk mengmbil data dari setiap baris (1 baris 1 koin)
    for row in rows:
        # Mengambil nama koin dan menyimpan data ke dalam list, jika data tidak ditemukan akan ditambahkan None ke dalam list
        try:
            coin = row.find('p', class_='sc-65e7f566-0 iPbTJf coin-item-name')
            listCoin.append(coin.get_text())
        except:
            listCoin.append(None)

        # Mengambil harga koin dan menyimpan data ke dalam list, jika data tidak ditemukan akan ditambahkan None ke dalam list
        try:
            priceDiv = row.find('div', class_='sc-b3fc6b7-0 dzgUIj')
            price = priceDiv.find('span')
            listPrice.append(price.get_text())
        except:
            listPrice.append(None)

        # Mengambil data market cap dan menyimpan data ke dalam list, jika data tidak ditemukan akan ditambahkan None ke dalam list
        try:
            marketcap = row.find('span', class_='sc-11478e5d-1 jfwGHx')
            listMarketCap.append(marketcap.get_text())
        except:
            listMarketCap.append(None)

        # Mengambil data volume dan menyimpan data ke dalam list, jika data tidak ditemukan akan ditambahkan None ke dalam list
        try:
            volume =  row.find('p', class_='sc-71024e3e-0 bbHOdE font_weight_500')
            listVolume.append(volume.get_text())

        except:
            listVolume.append(None)

        # Mengambil data coin supply dan menyimpan data ke dalam list, jika data tidak ditemukan akan ditambahkan None ke dalam list
        try:
            coinsupply = row.find('p', class_='sc-71024e3e-0 hhmVNu')
            listCoinSupply.append(coinsupply.get_text())
        except:
            listCoinSupply.append(None)

# Menutup browser
driver.close()
```

```python
# Membuat data frame
data = pd.DataFrame({'Coin Name':listCoin,
                     'Coin Price':listPrice,
                     'Market Cap':listMarketCap,
                     'Volume':listVolume,
                     'Supply':listCoinSupply
                     })
data
```

```python
# Mengecek missing value
data.info()
```

```python
# Menyimpan dataframe ke dalam csv files
data.to_csv("coinInfo.csv", index=False)
```

# Task 2 - Scrape Weather in 54 locations in Indonesia

Get weather data from 54 locations in Indonesia from https://www.timeanddate.com/weather/indonesia.

You should include the information of:

●Temperature

●Feels Like

●Wind Speed

●Wind Direction

●Humidity

●Dew Point

●Visibility

●Probability of Rain

●Amount of Rain-Only take the closest day and time (morning/evening) from the time you do this task

-Save into Pandas DataFrame and then save into excel or csv file5


```python
# Menyiapkan list kosong untuk menyimpan URL cuaca masing-masing kota
cities = []

# Membuka browser Chrome
driver = webdriver.Chrome()

# Membuka halaman utama dari world clock untuk Negara Indonesia
url = "https://www.timeanddate.com/worldclock/indonesia"
driver.get(url)

# Mengambil data html dari halaman
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Menemukan tabel yang berisi data kota yang memiliki link url
tabel = soup.find('table', class_="zebra fw tb-wc zebra")
body = tabel.find('tbody')
rows = body.find_all('tr')

# Looping untuk mengambil semua link dari tabel
for row in rows:
    try:
        data = row.find_all('a')
        for a in data:
            cities.append(a['href'])

    except:
        cities.append(None)

# Menutup browser
driver.close()
```

```python
# Mengkonversi link kota menjadi link cuaca dengan mengganti worldclock menjadi weather
links = []
for city in cities:
    url2 = f"https://www.timeanddate.com" + city.replace('worldclock','weather')
    links.append(url2)

# Menyiapkan list untuk menyimpan data semua kota
cityName = []
temperatures = []
feelLikes = []
windSpeeds = []
windDirections = []
humidities = []
dewPoints =[]
visibilities = []
probRains =[]
amountRains =[]

# Membuka halaman cuaca kota   
driver = webdriver.Chrome()

# Loop untuk setiap link cuaca dan mengambil data cuaca
for link in links:


    driver.get(link)

    # Menyimpan html halaman tiap kota
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    
    # Menemukan tabel cuaca di halaman web
    weatherTable = soup.find('table', id="wt-48")
    data = weatherTable.find_all('td')

    # Mengambil informasi nama cuaca dari tabel berdasarkan indeks karna tidak ditemukan class
    city = link.split("/")[len(link.split("/"))-1]
    temperature = data[7].get_text()
    feelLike = data[21].get_text()
    windSpeed = data[28].get_text()
    windDirection = data[35].get_text()
    humidity = data[42].get_text()
    dewPoint = data[49].get_text()
    visibility = data[56].get_text()
    probRain = data[63].get_text()
    amountRain = data[70].get_text()

    # Memasukkan infomasi yang di peroleh ke dalam list
    cityName.append(city)
    temperatures.append(temperature) 
    feelLikes.append(feelLike)
    windSpeeds.append(windSpeed)
    windDirections.append(windDirection)
    humidities.append(humidity)
    dewPoints.append(dewPoint)
    visibilities.append(visibility)
    probRains.append(probRain)
    amountRains.append(amountRain)

# Menutup browser
driver.close()
```

```python
# Membuat dataframe dari data yang sudah ada
data = pd.DataFrame({
    'City Name': cityName,
    'Temperature': temperatures,
    'Feels Like':feelLikes,
    'Wind Speed':windSpeeds,
    'Wind Direction':windDirections,
    'Humidity' : humidities,
    'Dew Point' : dewPoints,
    'Visibilities': visibilities,
    'Rain Probability': probRains,
    'Amount of Rain' : amountRains

})

# Merapikan nama kota karna diperoleh dari url
data['City Name'] = data['City Name'].str.replace('-'," ")
data['City Name'] = data['City Name'].str.capitalize()
```
